```python
import pandas as pd
import numpy as np
import random

# -------------------------------
# 1. DATE RANGE
# -------------------------------
dates = pd.date_range(start="2020-01-01", end="2025-12-31")

# -------------------------------
# 2. BLOOD TYPES
# -------------------------------
blood_types = ["A+", "A-", "B+", "B-", "AB+", "AB-", "O+", "O-"]

blood_weights = {
    "O+": 1.2, "B+": 1.1, "A+": 1.0,
    "AB+": 0.7, "O-": 0.4,
    "A-": 0.3, "B-": 0.3, "AB-": 0.2
}

# -------------------------------
# 3. HOSPITAL BASE DEMAND
# -------------------------------
hospital_base_demand = {
    "KEM Hospital": 95,
    "Lilavati Hospital": 70,
    "Kokilaben Dhirubhai Ambani Hospital": 75,
    "Lokmanya Tilak Municipal General Hospital": 85,
    "Tata Memorial Hospital": 90,
    "Breach Candy Hospital": 55,
    "Jaslok Hospital": 60,
    "Nanavati Max Super Speciality Hospital": 65,
    "Dr. L H Hiranandani Hospital": 60,
    "Bombay Hospital": 58,
    "Fortis Hospital Mulund": 62,
    "Global Hospital Mumbai": 70,
    "Wockhardt Hospitals Mumbai Central": 57,
    "Sir H. N. Reliance Foundation Hospital": 63,
    "P. D. Hinduja Hospital": 68
}

hospital_info = {
    "KEM Hospital": ("MUM-H-01", "Parel", 400012),
    "Lilavati Hospital": ("MUM-H-02", "Bandra", 400050),
    "Kokilaben Dhirubhai Ambani Hospital": ("MUM-H-03", "Andheri West", 400053),
    "Lokmanya Tilak Municipal General Hospital": ("MUM-H-04", "Sion", 400022),
    "Tata Memorial Hospital": ("MUM-H-05", "Parel", 400012),
    "Breach Candy Hospital": ("MUM-H-06", "Breach Candy", 400026),
    "Jaslok Hospital": ("MUM-H-07", "Pedder Road", 400026),
    "Nanavati Max Super Speciality Hospital": ("MUM-H-08", "Vile Parle", 400056),
    "Dr. L H Hiranandani Hospital": ("MUM-H-09", "Powai", 400076),
    "Bombay Hospital": ("MUM-H-10", "Marine Lines", 400020),
    "Fortis Hospital Mulund": ("MUM-H-11", "Mulund", 400080),
    "Global Hospital Mumbai": ("MUM-H-12", "Parel", 400012),
    "Wockhardt Hospitals Mumbai Central": ("MUM-H-13", "Mumbai Central", 400008),
    "Sir H. N. Reliance Foundation Hospital": ("MUM-H-14", "Girgaon", 400004),
    "P. D. Hinduja Hospital": ("MUM-H-15", "Mahim", 400016)
}

festival_days = [
    "2020-10-25", "2021-11-04", "2022-10-24",
    "2023-11-12", "2024-10-31", "2025-10-20"
]

rows = []

# -------------------------------
# 4. GENERATION LOOP
# -------------------------------
for date in dates:
    is_monsoon = 1 if date.month in [6, 7, 8, 9] else 0
    is_festival = 1 if str(date.date()) in festival_days else 0
    
    for hospital, info in hospital_info.items():
        hospital_id, location, pincode = info
        base = hospital_base_demand[hospital]
        
        for blood in blood_types:
            demand = base * blood_weights[blood]
            
            if is_monsoon:
                demand *= 1.2
            if is_festival:
                demand *= 1.15
            if date.weekday() >= 5:
                demand *= 0.95
            
            demand += np.random.normal(0, 6)
            demand = max(5, int(demand))
            
            rows.append([
                date.strftime("%d-%m-%Y"),
                hospital_id,
                hospital,
                location,
                pincode,
                blood,
                demand,
                date.strftime("%A"),
                is_monsoon,
                is_festival
            ])

# -------------------------------
# 5. CREATE DATAFRAME
# -------------------------------
columns = [
    "Date", "Hospital_ID", "Hospital_Name",
    "Hospital_Location", "Location_Pincode",
    "Blood_Type", "Units_Demanded",
    "Day_of_Week", "is_monsoon_peak",
    "is_major_festival"
]

df = pd.DataFrame(rows, columns=columns)

# -------------------------------
# 6. SAMPLE TO EXACT ~250K ROWS
# -------------------------------
df = df.sample(n=250000, random_state=42).reset_index(drop=True)

df.to_csv("synthetic_blood_demand_250k.csv", index=False)

print("Dataset generated with", len(df), "rows")
```
