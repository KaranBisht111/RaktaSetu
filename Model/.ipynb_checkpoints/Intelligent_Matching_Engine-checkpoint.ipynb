```python
# =================================================================
# RAKTASETU: INTELLIGENT MATCHING ENGINE (V3.0)
# Objective: Heuristic + AI Ranking of Donors
# =================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import os

# Visual Settings
%matplotlib inline
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print("Matching Engine Initialized")

# ---------------------------------------------------------------
# Load Data
# ---------------------------------------------------------------

donors = pd.read_csv("donors_with_predicted_availability.csv")
requests = pd.read_csv("synthetic_blood_matching_dataset.csv")

print("Total Donors:", len(donors))
print("Total Requests:", len(requests))
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(
    x='Patient_Blood_Type',
    data=requests,
    order=['A+', 'A-', 'B+', 'B-', 'O+', 'O-', 'AB+', 'AB-']
)
plt.title("Patient Requests by Blood Type")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(
    x='Blood_Type',
    data=donors,
    order=['A+', 'A-', 'B+', 'B-', 'O+', 'O-', 'AB+', 'AB-']
)
plt.title("Available Donors by Blood Type")
plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------------------
# Matching Engine Logic
# ---------------------------------------------------------------

class MatchingEngine:
    
    def __init__(self, donors_df):
        self.donors = donors_df.copy()

    def compatibility_score(self, patient_bt, donor_bt):
        if patient_bt == donor_bt:
            return 1.0

        rules = {
            "O+": ["O-"],
            "A+": ["A-", "O+", "O-"],
            "B+": ["B-", "O+", "O-"],
            "AB+": ["A+", "A-", "B+", "B-", "AB-", "O+", "O-"],
            "A-": ["O-"],
            "B-": ["O-"],
            "AB-": ["A-", "B-", "O-"],
            "O-": []
        }

        return 0.5 if donor_bt in rules.get(patient_bt, []) else 0.0

    def rank_donors(self, patient_bt, top_n=5):

        temp_df = self.donors.copy()

        temp_df["Comp_Score"] = temp_df["Blood_Type"].apply(
            lambda bt: self.compatibility_score(patient_bt, bt)
        )

        # Exact match priority
        eligible = temp_df[temp_df["Blood_Type"] == patient_bt].copy()

        if eligible.empty:
            eligible = temp_df[temp_df["Comp_Score"] > 0].copy()

        # Final weighted scoring
        eligible["Final_Score"] = (
            0.5 * eligible["Predicted_Availability"] +
            0.3 * eligible["Comp_Score"] +
            0.2 * (eligible["Total_Donations"] / eligible["Total_Donations"].max())
        )

        return eligible.sort_values("Final_Score", ascending=False).head(top_n)
```

```python
engine = MatchingEngine(donors)

os.makedirs("ml_models", exist_ok=True)
joblib.dump(engine, "ml_models/matching_engine.pkl")

print("Matching engine saved successfully.")
```

```python
results = []

for idx, row in requests.head(10).iterrows():

    top_5 = engine.rank_donors(row['Patient_Blood_Type'])

    for rank, d_row in enumerate(top_5.itertuples(), 1):
        results.append({
            "Request": row['Request_ID'],
            "Type": row['Patient_Blood_Type'],
            "Donor": d_row.Name,
            "Score": d_row.Final_Score,
            "Rank": rank
        })

match_results_df = pd.DataFrame(results)

match_results_df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=match_results_df,
    x='Request',
    y='Score',
    hue='Rank'
)

plt.title("Top 5 Ranked Donors per Request")
plt.ylabel("Match Confidence Score")
plt.legend(title="Rank", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(match_results_df['Score'], bins=20, kde=True)
plt.title("Distribution of Match Confidence Scores")
plt.tight_layout()
plt.show()
```

```python
match_results_df.to_csv("final_ranked_matches.csv", index=False)

avg_score = match_results_df['Score'].mean()

print("Matching Completed.")
print("Average Match Confidence:", round(avg_score, 2))
```

```python
# =================================================================
# RAKTASETU: INTELLIGENT MATCHING ENGINE (V3.0)
# Objective: Heuristic Ranking of Donors for Patient Requests
# =================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import os

# Visual configuration
%matplotlib inline
sns.set_theme(style="white", palette="muted")
plt.rcParams['figure.figsize'] = (12, 6)

print("✅ Integration Initialized: Ready to link Donors with Patient Requests.")
```

```python
# ==============================================================
# 📊 Donor Blood Type Distribution
# ==============================================================

plt.figure(figsize=(8, 5))
sns.countplot(x='Blood_Type', data=donors,
              order=['A+', 'A-', 'B+', 'B-', 'O+', 'O-', 'AB+', 'AB-'])
plt.title("Available Donors by Blood Type")
plt.tight_layout()
plt.show()
```

```python
# 1. Load Data
donors = pd.read_csv("donors_with_predicted_availability.csv")
requests = pd.read_csv("synthetic_blood_matching_dataset.csv")

# 2. Visual Check: Request Volume by Blood Type
plt.figure(figsize=(10, 5))
sns.countplot(x='Patient_Blood_Type', data=requests, palette='Reds_r', order=['A+', 'A-', 'B+', 'B-', 'O+', 'O-', 'AB+', 'AB-'])
plt.title("Current Mumbai Hospital Requests by Blood Type", fontsize=14)
plt.ylabel("Number of Patients Waiting")
plt.show()

print(f"Total Available Donors: {len(donors)}")
print(f"Total Pending Requests: {len(requests)}")
```

```python
class MatchingEngine:
    def __init__(self, donors_df):
        self.donors = donors_df.copy()

    def compatibility_score(self, patient_bt, donor_bt):
        # Universal compatibility logic (ABO rules)
        # 1.0 = Exact Match | 0.5 = Compatible | 0 = Incompatible
        if patient_bt == donor_bt: return 1.0
        
        rules = {
            "O+": ["O-"],
            "A+": ["A-", "O+", "O-"],
            "B+": ["B-", "O+", "O-"],
            "AB+": ["A+", "A-", "B+", "B-", "AB-", "O+", "O-"],
            "A-": ["O-"], "B-": ["O-"], "AB-": ["A-", "B-", "O-"], "O-": []
        }
        return 0.5 if donor_bt in rules.get(patient_bt, []) else 0.0

    def rank_donors(self, patient_bt, top_n=5):
        temp_df = self.donors.copy()
        
        # Apply Logic: Score = (50% AI Prob) + (30% Compatibility) + (20% Donation History)
        temp_df["Comp_Score"] = temp_df["Blood_Type"].apply(lambda bt: self.compatibility_score(patient_bt, bt))
        
        # Filter only those who can actually give blood to this patient
        eligible = temp_df[temp_df["Blood_Type"] == patient_bt].copy() # Priority to exact match
        if eligible.empty: # Fallback to compatible types
            eligible = temp_df[temp_df["Comp_Score"] > 0].copy()

        eligible["Final_Score"] = (
            0.5 * eligible["Predicted_Availability"] + 
            0.3 * eligible["Comp_Score"] + 
            0.2 * (eligible["Total_Donations"] / eligible["Total_Donations"].max())
        )

        return eligible.sort_values("Final_Score", ascending=False).head(top_n)

# Initialize and Save
engine = MatchingEngine(donors)
joblib.dump(engine, "ml_models/matching_engine.pkl")
print("🚀 Matching Engine Brain saved to ml_models/matching_engine.pkl")
```

```python
# Run matches for the current requests
results = []
for idx, row in requests.head(10).iterrows():
    top_5 = engine.rank_donors(row['Patient_Blood_Type'])
    for rank, d_row in enumerate(top_5.itertuples(), 1):
        results.append({
            "Request": row['Request_ID'],
            "Type": row['Patient_Blood_Type'],
            "Donor": d_row.Name,
            "Score": d_row.Final_Score,
            "Rank": rank
        })

match_results_df = pd.DataFrame(results)

# Visualizing the Match Priority
plt.figure(figsize=(12, 6))
sns.barplot(data=match_results_df, x='Request', y='Score', hue='Rank', palette='viridis')
plt.title("Intelligent Matching Engine: Top 5 Ranked Donors per Request", fontsize=15)
plt.ylabel("Match Confidence Score (0-1.0)")
plt.legend(title="Donor Rank", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Save the final ranked matches to CSV
match_results_df.to_csv("final_ranked_matches.csv", index=False)

# Final Statistical Insight
avg_score = match_results_df['Score'].mean()
print(f"✅ Success: Analysis Complete.")
print(f"Average Match Confidence for Mumbai Sector: {avg_score:.2f}")
print("The 'matching_engine.pkl' is now ready to be loaded into Flask/app.py.")
```
